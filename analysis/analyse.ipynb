```python
from collections import Counter
import operator

import matplotlib.pyplot as plt
%matplotlib inline

import sys
sys.path.append('..')
from src.entities.dataset import BucketDataset, Dataset
from src.data.loaders import DataLoader
```

```python
def show_dataset_statistics(dataset: Dataset):
    print('Dataset size:', len(dataset))
    for sample in dataset[:10]:
        print(sample)
    
    f, ax = plt.subplots(ncols=4)
    f.set_figheight(3)
    f.set_figwidth(20)
    f.dpi = 150
    
    # Word lengths
    n, bins, patches = ax[0].hist(x=[len(sample.word) for sample in dataset], bins=50, facecolor='b')
    ax[0].set_xlabel('Word lengths')
    ax[0].set_ylabel('Counts')
    
    # Number of segments for each word 
    n, bins, patches = ax[1].hist(x=[len(sample.segments) for sample in dataset], bins=30, facecolor='r')
    ax[1].set_xlabel('Number of segments for each word')
    ax[1].set_ylabel('Counts')
    
    
    # Occurances of word segment parts
    segments = Counter([seg for sample in dataset for seg in sample.segment_parts])
    sorted_segments = sorted(segments.items(), key=operator.itemgetter(1))[::-1]

    ax[2].scatter(x=list(range(len(segments))), y=[seg[1] for seg in sorted_segments])
    ax[2].set_xlabel('Occurances of word segment parts')
    ax[2].set_ylabel('Counts')

    sorted_segments[:10]
    
    plt.subplot(1, 4, 4)
    n, bins, patches = ax[3].hist(x=[len(seg) for sample in dataset for seg in sample.segment_parts], bins=15, facecolor='y')
    ax[3].set_xlabel('Number of segments for each word')
    ax[3].set_ylabel('Counts')
    
    plt.show()
```

```python
show_dataset_statistics(dataset=Dataset(samples=DataLoader(file_path='../datasets/rus.train').load()))
```
